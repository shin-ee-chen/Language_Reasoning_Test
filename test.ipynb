```python
from PIL import Image
import matplotlib.pyplot as plt
import utils_show as utils

img_path = utils.get_img_path(2354915)

def show_image(img_path, show=False):
    image = Image.open(img_path).convert("RGB")
    if show:
        plt.imshow(image)
        plt.axis('off')
        plt.show()
    return image
show_image(img_path, show=False)
```
